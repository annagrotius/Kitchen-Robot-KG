```python
import rdflib
```

```python
KC_aff = './graphs/aff_bft_open_graph.ttl'
KC_aff_cn = 'aff_bft_cn_Final.ttl'
# KC_aff_cn_puppy = 
```

```python
KC_aff_g = rdflib.Graph()
result = KC_aff_g.parse(KC_aff, format='turtle')
len(result)
```

```python
relations = []
for p in KC_aff_g.predicates():
    relations.append(p)
print(len(set(relations)))
```

```python
entities = []
for s,p,o in KC_aff_g.triples((None,None,None)):
    entities.append(s)
    entities.append(o)
print(len(set(entities)))
```

```python
for i in entities:
    print(type(i))
    if type(i) == '<class rdflib.term.BNode>':
        print(i)
        entities.remove(i)
print(len(set(entities)))
```
